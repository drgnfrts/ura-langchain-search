## Unstructured.io Quick Start

I've also included some a quick script to demonstrate how you might want to use unstructured io's scraping function. 
Do remember to use the headers to fool the site into giving you data or it wouldn't work. 
Have fun! 

```python
from unstructured.partition.html import partition_html

# Using partition_html to ingest HTML content
input_html = "https://www.ura.gov.sg/Corporate/Media-Room/Media-Releases/pr24-03"
document_elements = partition_html(url=input_html,
                                   headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'})
document_elements
```
